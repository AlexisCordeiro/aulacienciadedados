```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
```

```python
base_df = pd.read_csv("/content/drive/MyDrive/base_CIDA/indexData.csv")
base_df
```

```python
datadf1 = base_df[["Index", "Close", "Date"]]
datadf1
```

```python
dfremove = datadf1.loc[(datadf1["Index"]!="NYA")]
datanyadf = datadf1.drop(dfremove.index)
datanyadf = datanyadf[13490:]
datanyadf
```

```python
figura = plt.figure(figsize=(15,5))
plt.plot(datanyadf["Date"],datanyadf["Close"], label = "NYA", color = "red", ls = '--', lw = 2)
plt.legend(loc = 2, fontsize = 16)
plt.xlabel("Dia")
plt.ylabel("Cotação")
plt.title("Cotação da Bolsa de NY")
plt.show(figura)
```
